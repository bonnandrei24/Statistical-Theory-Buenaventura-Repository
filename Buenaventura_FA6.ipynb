```python
# ============================================================
# FA6 – Statistical Theory
# Solutions using Python
# Name: Bonn Andrei Buenaventura
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import binom
from itertools import product

# ============================================================
# PROBLEM 8.18
# ============================================================

print("\n================ PROBLEM 8.18 ================\n")
# Problem 8.18 – Sampling Distribution Verification

import numpy as np

# 1. Population Setup
population = [9, 12, 15]
p = 1/3

# Calculation using your logic but refined with f-strings
mu = sum(x * p for x in population)
var = sum((x**2) * p for x in population) - mu**2

print(f"Population mean (μ): {mu}")
print(f"Population variance (σ²): {var}")
print("-" * 30)

# 2. Generating Sample Means (List Comprehension is faster/cleaner)
# This replicates the 'with replacement' requirement (3x3 = 9 samples)
sample_means = [(x + y) / 2 for x in population for y in population]

# 3. Sampling Distribution Calculations
mu_xbar = sum(sample_means) / len(sample_means)
var_xbar = (sum(m**2 for m in sample_means) / len(sample_means)) - mu_xbar**2

# 4. Formal Proof Output
print(f"Mean of sampling distribution (μ_x̄): {mu_xbar}")
print(f"Variance of sampling distribution (σ²_x̄): {var_xbar}")
print("\n--- FORMAL PROOF ---")

# We use np.isclose in case of tiny floating point math errors
if np.isclose(mu_xbar, mu):
    print(f"Property 1 Verified: μ_x̄ ({mu_xbar}) is equal to μ ({mu}).")

if np.isclose(var_xbar, var / 2):
    print(f"Property 2 Verified: σ²_x̄ ({var_xbar}) is equal to σ²/n ({var/2}).")

# ============================================================
# PROBLEM 8.21
# ============================================================

print("\n================ PROBLEM 8.21 ================\n")

population = np.array([3,7,11,15])
n = 2

# Population parameters
mu = np.mean(population)
sigma = np.std(population)

print("Population Mean (μ):", mu)
print("Population Standard Deviation (σ):", sigma)

# Generate samples
samples = list(product(population, repeat=n))
sample_means = [np.mean(s) for s in samples]

df_821 = pd.DataFrame({
    "Sample": samples,
    "Sample Mean": sample_means
})

print("\nAll Possible Samples:")
print(df_821)

# Sampling distribution
sampling_dist = df_821["Sample Mean"].value_counts(normalize=True).sort_index()

print("\nSampling Distribution of Sample Mean:")
print(sampling_dist)

# Statistics
mean_sampling = np.mean(sample_means)
std_sampling = np.std(sample_means)

print("\nMean of Sampling Distribution:", mean_sampling)
print("Standard Deviation of Sampling Distribution:", std_sampling)

# Formula verification
std_formula = sigma / np.sqrt(n)

print("\nVerification using formula σ/√n =", std_formula)


# ============================================================
# PROBLEM 8.34
# ============================================================

print("\n================ PROBLEM 8.34 ================\n")

n = 200
p = 0.5

# (a) Less than 40% boys
boys_40 = int(0.40 * 200)
prob_a = binom.cdf(boys_40 - 1, n, p)

# (b) Between 43% and 57%
low = int(0.43 * 200)
high = int(0.57 * 200)
prob_b = binom.cdf(high, n, p) - binom.cdf(low - 1, n, p)

# (c) More than 54% boys
boys_54 = int(0.54 * 200)
prob_c = 1 - binom.cdf(boys_54, n, p)

print("(a) P(Less than 40% boys) =", round(prob_a,4))
print("(b) P(43% to 57%) =", round(prob_b,4))
print("(c) P(More than 54% boys) =", round(prob_c,4))


# ============================================================
# PROBLEM 8.49
# ============================================================

print("\n================ PROBLEM 8.49 ================\n")

x = np.array([6,9,12,15,18])
p = np.array([0.1,0.2,0.4,0.2,0.1])

# Population mean
mu = np.sum(x*p)

# Population variance
variance = np.sum((x-mu)**2 * p)

print("Population Mean (μ):", mu)
print("Population Variance (σ²):", variance)

# All samples size 2
samples = list(product(x, repeat=2))

means = []
probs = []

for a,b in samples:

    mean_val = (a+b)/2

    pa = p[np.where(x==a)][0]
    pb = p[np.where(x==b)][0]

    prob = pa * pb

    means.append(mean_val)
    probs.append(prob)

df_849 = pd.DataFrame({
    "Sample": samples,
    "Sample Mean": means,
    "Probability": probs
})

print("\nAll 25 Possible Samples:")
print(df_849)

# Sampling distribution statistics
mean_sampling = np.sum(df_849["Sample Mean"] * df_849["Probability"])
var_sampling = np.sum((df_849["Sample Mean"] - mean_sampling)**2 * df_849["Probability"])

print("\nMean of Sampling Distribution:", mean_sampling)
print("Variance of Sampling Distribution:", var_sampling)
```
