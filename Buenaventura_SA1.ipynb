```python
# ============================================
# STATISTICAL THEORY - PYTHON SOLUTIONS
# ============================================

import pandas as pd
import numpy as np
from scipy import stats

title = "Statistical Theory Summative Assessment"
name  = "Bonn Andrei M. Buenaventura"

print("\n" + "="*70)
print(f"{title:^70}")
print(f"{name:^70}")
print("="*70)

# ============================================
# DATA PREPARATION
# ============================================

males = [
12, 4, 11, 13, 11,
7, 9, 10, 10, 7,
7, 12, 6, 9, 15,
10, 11, 12, 7, 8,
8, 9, 11, 10, 9,
10, 9, 9, 7, 9,
11, 7, 10, 10, 11,
9, 12, 12, 8, 13,
9, 10, 8, 11, 10,
13, 13, 9, 10, 13
]

females = [
11, 9, 7, 10, 9,
10, 10, 7, 9, 10,
11, 8, 9, 6, 11,
10, 7, 9, 12, 14,
11, 12, 12, 8, 12,
12, 9, 10, 11, 7,
12, 7, 9, 8, 11,
10, 8, 13, 8, 10,
9, 9, 9, 11, 9,
9, 8, 9, 12, 11
]

df = pd.DataFrame({
    'Hours': males + females,
    'Gender': ['Male']*50 + ['Female']*50
})

# ============================================
# FUNCTION: DESCRIPTIVE STATISTICS
# ============================================

def get_stats(data):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)
    ci_low, ci_high = stats.t.interval(0.95, df=n-1, loc=mean, scale=std_err)

    results = {
        "Mean": mean,
        "Median": np.median(data),
        "Std Dev": np.std(data, ddof=1),
        "Variance": np.var(data, ddof=1),
        "Min": np.min(data),
        "Max": np.max(data),
        "Sample Size": n,
        "Skewness": stats.skew(data, bias=False),
        "Kurtosis": stats.kurtosis(data, bias=False),
        "Q1 (25%)": np.percentile(data, 25),
        "Q2 (50%)": np.percentile(data, 50),
        "Q3 (75%)": np.percentile(data, 75),
        "95% CI Lower": ci_low,
        "95% CI Upper": ci_high
    }

    for i in range(1, 10):
        results[f"D{i} (P{i*10})"] = np.percentile(data, i*10)

    return pd.Series(results)

# ============================================
# PROBLEM I — DESCRIPTIVE STATISTICS
# ============================================

def format_stats_table(stats_series):
    """Format statistics for cleaner display"""
    stats_df = stats_series.to_frame(name="Value")
    return stats_df.round(4)

# --------------------------------------------
# PROBLEM I.1 — ENTIRE DATASET
# --------------------------------------------
print("\n" + "="*60)
print("PROBLEM I.1 — DESCRIPTIVE STATISTICS (ENTIRE DATASET)")
print("="*60)

overall_stats = get_stats(df['Hours'])
display(format_stats_table(overall_stats))

# --------------------------------------------
# PROBLEM I.2 — GROUPED BY GENDER
# --------------------------------------------
print("\n" + "="*60)
print("PROBLEM I.2 — DESCRIPTIVE STATISTICS (BY GENDER)")
print("="*60)

grouped_stats = df.groupby('Gender')['Hours'].apply(get_stats).unstack().T
display(grouped_stats.round(4))

# ============================================
# PROBLEM I.3 — STATISTICAL REPORT
# ============================================

print("\n" + "="*60)
print("PROBLEM I.3 — STATISTICAL REPORT")
print("="*60)

# Extract key values
overall_mean = overall_stats["Mean"]
overall_median = overall_stats["Median"]
overall_std = overall_stats["Std Dev"]
overall_ci_low = overall_stats["95% CI Lower"]
overall_ci_high = overall_stats["95% CI Upper"]

male_stats = grouped_stats["Male"]
female_stats = grouped_stats["Female"]

# Report
print(f"""
STATISTICAL REPORT

1. Descriptive Statistics (Overall Dataset)
The dataset has a mean of {overall_mean:.2f} hours and a median of {overall_median:.2f} hours,
indicating a fairly symmetric distribution. The standard deviation is {overall_std:.2f},
suggesting moderate variability in the number of hours. The values range from
{overall_stats["Min"]:.0f} to {overall_stats["Max"]:.0f} hours.

2. Comparison by Gender
Male students have a mean of {male_stats["Mean"]:.2f} hours, while female students
have a mean of {female_stats["Mean"]:.2f} hours. The variability is slightly
{"higher" if male_stats["Std Dev"] > female_stats["Std Dev"] else "lower"} among males
(std = {male_stats["Std Dev"]:.2f}) compared to females (std = {female_stats["Std Dev"]:.2f}).

3. 95% Confidence Intervals for the Mean
The overall 95% confidence interval for the mean is ({overall_ci_low:.2f}, {overall_ci_high:.2f}),
which means we are 95% confident that the true average hours lies within this range.

For males, the 95% CI is ({male_stats["95% CI Lower"]:.2f}, {male_stats["95% CI Upper"]:.2f}),
while for females, it is ({female_stats["95% CI Lower"]:.2f}, {female_stats["95% CI Upper"]:.2f}).
The overlap of these intervals suggests that the difference in means may not be statistically significant.

4. Observed Patterns and Differences
Both groups show similar central tendencies, with only slight differences in mean values.
The distributions appear fairly balanced, and no extreme skewness is observed.
Overall, gender does not appear to create a substantial difference in hours based on this dataset.
""")
# ============================================
# PROBLEM II — MOMENTS
# ============================================

print("\n=== PROBLEM II: MOMENTS ===")

data_set = np.array([2, 3, 7, 8, 10, 10, 12, 15, 17, 20])
mean_val = np.mean(data_set)

# II.1 Raw Moments (about zero)
m1p = np.mean(data_set**1)
m2p = np.mean(data_set**2)
m3p = np.mean(data_set**3)
m4p = np.mean(data_set**4)

print(f"\nII.1 Raw Moments")
print(f"m1' = {m1p:.4f}")
print(f"m2' = {m2p:.4f}")
print(f"m3' = {m3p:.4f}")
print(f"m4' = {m4p:.4f}")

# II.2 Central Moments (about the mean)
m1 = np.mean((data_set - mean_val)**1)
m2 = np.mean((data_set - mean_val)**2)
m3 = np.mean((data_set - mean_val)**3)
m4 = np.mean((data_set - mean_val)**4)

print(f"\nII.2 Central Moments")
print(f"m1 = {m1:.4f}")
print(f"m2 = {m2:.4f}")
print(f"m3 = {m3:.4f}")
print(f"m4 = {m4:.4f}")

# II.3 Verification Formula
rhs = m4p - 4*m1p*m3p + 6*(m1p**2)*m2p - 3*(m1p**4)

print(f"\nII.3 Verification")
print(f"Central m4        = {m4:.4f}")
print(f"Formula Result    = {rhs:.4f}")
print(f"Match: {np.isclose(m4, rhs)}")

# ============================================
# PROBLEM III — MOMENT RELATIONSHIP PROOF
# ============================================

print("\n" + "="*60)
print("PROBLEM III — VERIFY MOMENT RELATIONSHIP")
print("="*60)

print("\nGiven:")
print("m4' = m4 + 4μm3 + 6μ²m2 + μ⁴")
print(f"Mean (μ) = m1' = {m1p:.4f}")
print(f"Central Moments: m2={m2:.4f}, m3={m3:.4f}, m4={m4:.4f}")
print(f"Raw 4th Moment (m4') = {m4p:.4f}")

# Step 1 — Compute each term
term1 = m4
term2 = 4*m1p*m3
term3 = 6*(m1p**2)*m2
term4 = m1p**4

print("\nStep 1 — Compute RHS Terms")
print(f"m4              = {term1:.4f}")
print(f"4μm3            = {term2:.4f}")
print(f"6μ²m2           = {term3:.4f}")
print(f"μ⁴              = {term4:.4f}")

# Step 2 — Sum RHS
rhs_proof = term1 + term2 + term3 + term4

print("\nStep 2 — Sum RHS")
print(f"RHS = {rhs_proof:.4f}")

# Step 3 — Compare
print("\nStep 3 — Compare LHS and RHS")
print(f"LHS (m4') = {m4p:.4f}")
print(f"RHS       = {rhs_proof:.4f}")
print(f"Proof Holds: {np.isclose(m4p, rhs_proof)}")

# ============================================
# PROBLEM IV — STUDENT'S t DISTRIBUTION
# ============================================

print("\n" + "="*60)
print("PROBLEM IV — t-DISTRIBUTION CRITICAL VALUES")
print("="*60)

# IV.1
print("\nIV.1 Area between -t and t is 0.90, df = 20")
print("Total middle area = 0.90")
print("Remaining area (α) = 1 - 0.90 = 0.10")
print("Two tails → α/2 = 0.05")
print("Left cumulative = 1 - 0.05 = 0.95")

t1_1 = stats.t.ppf(0.95, df=20)
print(f"t = {t1_1:.4f}")

# IV.2
print("\nIV.2 Area left of -t is 0.025, df = 20")
print("So -t is 2.5th percentile")

t1_2 = -stats.t.ppf(0.025, df=20)
print(f"t = {t1_2:.4f}")

# IV.3
print("\nIV.3 Combined tails area = 0.01, df = 5")
print("Total α = 0.01")
print("Each tail = 0.005")
print("Left cumulative = 1 - 0.005 = 0.995")

t1_3 = stats.t.ppf(0.995, df=5)
print(f"t = {t1_3:.4f}")

# IV.4
print("\nIV.4 Area right of t is 0.55, df = 16")
print("Right tail = 0.55")
print("Left cumulative = 1 - 0.55 = 0.45")

t1_4 = stats.t.ppf(0.45, df=16)
print(f"t = {t1_4:.4f}")

#print("\n" + "="*50)
#print("ALL PROBLEMS SOLVED SUCCESSFULLY")
#print("="*50)
```
