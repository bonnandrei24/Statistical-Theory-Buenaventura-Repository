```python
import numpy as np
import pandas as pd

# DATA

normal = [
67,69,70,62,63,67,65,59,68,66,60,65,70,63,64,65,69,60,61,67,
66,64,65,68,71,61,62,69,66,65,68,62,64,67,67,70,62,64,66,63,
65,68,63,64,66,65,65,61,63,66
]

skewed_right = [
31,40,43,24,30,29,30,24,38,27,26,35,29,33,55,75,46,38,26,34,
29,85,57,29,34,40,34,41,36,35,40,26,28,34,26,19,66,23,63,28,
30,26,33,31,24,25,35,22,34,28
]

skewed_left = [
102,87,55,104,70,75,95,80,73,66,79,93,60,90,73,84,89,73,85,98,
72,79,92,35,76,71,93,90,76,71,97,63,10,58,70,82,85,72,25,93,
83,44,58,65,10,77,92,81,82,77
]

uniform = [
12.1,11.6,12.1,11.6,12.4,12.0,12.1,11.6,12.1,11.6,
12.2,11.7,12.2,12.3,12.2,11.7,11.9,11.7,12.2,11.7,
12.3,11.8,12.3,12.5,11.7,11.8,12.3,11.8,12.3,11.8,
12.4,11.9,12.4,11.9,12.1,11.9,12.4,12.2,12.4,11.9,
12.5,12.0,11.8,11.9,12.5,12.0,12.5,12.0,12.5,12.0
]

datasets = {
    "Normal": normal,
    "Skewed Right": skewed_right,
    "Skewed Left": skewed_left,
    "Uniform": uniform
}

# COMPUTE RAW MOMENTS (about origin)

def raw_moments(data, k):
    data = np.array(data)
    return np.mean(data**k)

# COMPUTE CENTRAL MOMENTS (about mean)

def central_moments(data, k):
    data = np.array(data)
    mean = np.mean(data)
    return np.mean((data - mean)**k)

# ------------------------------------------------
# Question 1: Find the (a) first, (b) second, (c) third, and (d ) fourth moments
# For each of the sets of data (normal, skewed-right, skewed-left, uniform)
# ------------------------------------------------
print("QUESTION #1 ABOUT ORGIN\n")

for name, data in datasets.items():
    print(f"{name}")
    for k in range(1,5):
        print(f"Moment {k}' =", raw_moments(data, k))
    print()

# ------------------------------------------------
# Question 2: Find the (a) first, (b) second, (c) third, and (d ) fourth moments
# About the mean for each of the sets of data (normal, skewed-right, skewed-left, uniform).
# ------------------------------------------------
print("\nQUESTION #2 ABOUT MEAN\n")

for name, data in datasets.items():
    print(f"{name}")
    for k in range(1,5):
        print(f"Moment {k} =", central_moments(data, k))
    print()

# ------------------------------------------------
# Question 3: Find the (a) first, (b) second, (c) third, and (d ) fourth moments
# About the number 75 for the set of female height measurements.
# ------------------------------------------------
print("\nQUESTION #3 MOMENTS ABOUT 75\n")

normal_array = np.array(normal)

for k in range(1,5):
    moment_75 = np.mean((normal_array - 75)**k)
    print(f"Moment {k} about 75 =", moment_75)

# ------------------------------------------------
# Question 4: Using the results of items 2 and 3 for the set of female height measurements
# Verify the relations between the moments
# ------------------------------------------------
print("\nQUESTION #4 VERIFY THE RELATIONS\n")

m1p = raw_moments(normal,1)
m2p = raw_moments(normal,2)
m3p = raw_moments(normal,3)
m4p = raw_moments(normal,4)

m2 = central_moments(normal,2)
m3 = central_moments(normal,3)
m4 = central_moments(normal,4)

print("Moment 2 = {:.6f}  &  m2p - m1p**2 = {:.6f}".format(
    m2, m2p - m1p**2))

print("Moment 3 = {:.6f}  &  m3p - 3*m1p*m2p + 2*(m1p**3) = {:.6f}".format(
    m3, m3p - 3*m1p*m2p + 2*(m1p**3)))

print("Moment 4 = {:.6f}  &  m4p - 4*m1p*m3p + 6*(m1p**2)*m2p - 3*(m1p**4) = {:.6f}".format(
    m4, m4p - 4*m1p*m3p + 6*(m1p**2)*m2p - 3*(m1p**4)))
```
